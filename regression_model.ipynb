```python
import matplotlib.pyplot as plt
import numpy as np
#balicky
```

```python
def regression(x1_area,x2_beds,b1 = 148.73,b2 = -18.85,v = 54.4):
    prediction = b1*x1_area+b2*x2_beds+v
    return prediction
#funkce
```

```python
x = np.linspace(0,800,100)
y=x
plt.plot(x,y, linestyle='--', color='red',label='the perfect prediction line')

actual_price = np.array([115.0,234.5,198.0,528.0,572.5])

x1_area = np.array([0.846,1.324,1.150,3.037,3.984])
x2_beds = np.array([1,2,3,4,5])
colours = ['b','g','y','r', 'm']

predicted_price = regression(x1_area,x2_beds)

for i in range(len(actual_price)):
    plt.scatter(actual_price[i], predicted_price[i], color=colours[i],label=f'price of house {i+1}')
    plt.text(actual_price[i] + 10,
             predicted_price[i] + 10,
             f'House {i+1}',
             fontsize=9)
#to fialove f s uvoyovkami je prom2nn8 v textu

plt.title('Graph of actual price vs predicted price for 5 houses')
plt.xlabel('Actual price')
plt.ylabel('Pdredicted price')
plt.legend()
plt.xlim(left=0, right=800)
plt.ylim(bottom=0, top=800)

plt.show()
```
